#### Cleaning Data Project

```python
import pandas as pd
```

```python
df = pd.read_csv(r'E:\Learning_Files\bank.csv')
df
```

```python
# An excellent first step is to drop duplicates in the dataset, this could eliminate "low hanging fruit" on the front end of the cleaning
df = df.drop_duplicates()
df
```

```python
# As we can see, there weren't any duplicates in this set, but that's still a good check to perform
# Having looked at this Data Set in Excel, it seems that everything is stacked in one column, we can look up the shape to confirm
df.shape
```

```python
# Hypothesis confirmed.  Definitely time to split this thing up into columns, and luckily there's a consistent delimiter of ;
# Initial count is 16 ';' so we'll try to split this up on that and see how it goes
df[["age","job","marital","education","default","balance","housing","loan","contact","day","month","duration","campaign","pdays","previous","poutcome","y"]] = df['age;"job";"marital";"education";"default";"balance";"housing";"loan";"contact";"day";"month";"duration";"campaign";"pdays";"previous";"poutcome";"y"'].str.split(';', n = 16, expand=True)
df
```

```python
# So!  We've broken up all of the data that was separated by the ';'.  In doing what we did, we kept the original column as well, so we can drop that.
df.drop(columns=['age;"job";"marital";"education";"default";"balance";"housing";"loan";"contact";"day";"month";"duration";"campaign";"pdays";"previous";"poutcome";"y"'], inplace=True)
df
```

```python
# This has left us with a parseable DataFrame, and we're able to now work on the formatting of the various data points.
# First, we should drop out all the " surrounding the various data entries
df = df.replace(r'"','', regex=True)
df
```

```python
# Now we have a fully cleaned DataFrame.
# Let's look at the final column, 'y', to see if there are any other values than no, if that's the case we can probably just eliminate that column entirely
df['y'].unique()
```

```python
# So, there *are* some yes's that exist in this column, for now we should leave them alone, or ask the stakeholder the purpose of that cloumn to see if it's necessary or something we want to filter
# Now that we've created a parseable list, with no further instructions from the stakeholder, we could provide this to the user
df.to_csv('bank_cleaned.csv', index=False)
```
